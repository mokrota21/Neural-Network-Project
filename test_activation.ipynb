```python
def y(x):
    return np.dot([1, 2, 3], x) + 1
```

```python
def test_samples(size):
    x_set = []
    y_set = []
    for _ in range(size):
        x = np.random.random(3)
        x_set.append(x.reshape(-1, 1))
        y_set.append(y(x).reshape(-1, 1))
    return x_set, y_set
```

```python
network = NeuralNetwork([
    DenseLayer(3, 3),
    ActivationLayer(),
    DenseLayer(3, 1)
], 0.01)
```

```python
epochs = 1000
size = 20
x_set, y_set = test_samples(size)
for epoch in range(epochs):
    print(f"Epoch: {epoch}")
    network.train(x_set, y_set)
```

```python
x_test_set, y_test_set = test_samples(10)
for i in range(10):
    print(f"Predicted: {network.predict(x_test_set[i])}, Expected: {y_test_set[i]}")
```
