```python
from network import NeuralNetwork
from dense import DenseLayer
```

```python
x = np.random.random((3))
y = np.random.random((3))

print(f"x: {x}'\ny: {y}\nx * y: {np.dot(x, y).reshape(-1, 1)}\nx * y + 1: {np.dot(x, y).reshape(-1, 1) + 1}")
```

```python
def y(x):
    return np.dot([1, 2], x) + 1
```

```python
def test_samples(size):
    x_set = []
    y_set = []
    for _ in range(size):
        x = np.random.random(2)
        x_set.append(x.reshape(-1, 1))
        y_set.append(y(x).reshape(-1, 1))
    return x_set, y_set
```

```python
x = np.random.random((5, 1))
print(x)
s = 0
for i in range(x.shape[0]):
    s += x[i, 0] ** 2
print((x * x).sum(), s)
```

```python
network = NeuralNetwork([
    DenseLayer(2, 3),
    DenseLayer(3, 1)
], 0.01)

epochs = 1000
size = 9
x_set, y_set = test_samples(size)
for epoch in range(epochs):
    print(f"Epoch: {epoch}")
    network.train(x_set, y_set)
```

```python
for layer in network.layers_:
    print(layer.w_)
```

```python
x_test_set, y_test_set = test_samples(10)
for i in range(10):
    print(f"Predicted: {network.predict(x_test_set[i])}, Expected: {y_test_set[i]}")
```
